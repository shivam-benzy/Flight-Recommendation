```python
import random
import pandas as pd
import datetime
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
# from annoy import AnnoyIndex

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
route = ['New York (JFK) - Los Angeles (LAX)', 'Chicago (ORD) - Las Vegas (LAS)', 'Houston (IAH) - Denver (DEN)', 'Miami (MIA) - New York (JFK)', 'Boston (BOS) - Orlando (MCO)', 'Las Vegas (LAS) - Los Angeles (LAX)', 'Seattle (SEA) - San Francisco (SFO)', 'Denver (DEN) - Portland (PDX)', 'Phoenix (PHX) - Dallas (DFW)', 'San Francisco (SFO) - Seattle (SEA)']
airline=['United Airlines','JetBlue Airways','Delta Air Lines','Southwest Airlines']
class1=["Economy","First","Business"]
random_seed = 42
random.seed(random_seed)
```

```python
st=[]
mp4={}
mp={}
mp2=[]

for i in range(5000):
    route1=random.choice(route)
    mp2.append(route1)
    airline1=random.choice(airline)
    flight_no=random.randint(0,9)
    flight_id=airline1[0:2]+str(flight_no)+":"+route1+":"+airline1
    if flight_id not in st:
        st.append(flight_id)

        k=0
        for keys in mp4.keys():
            for j in mp4[keys]:
                if flight_id == j:
                    k=1
        
        if k==0:
            if route1 not in mp4:
                mp4[route1] = [flight_id]
            else:
                mp4[route1].append(flight_id)
            start_hour = random.randint(0, 22)
            minute = random.randint(0, 59)
            second = random.randint(0, 59)
            start_time = datetime.time(start_hour, minute, second)

            end_hour = random.randint(start_hour, 23) 
            minute = random.randint(0, 59)
            second = random.randint(0, 59)
            end_time= datetime.time(end_hour, minute, second)
        
            mp[flight_id]=[start_time,end_time]
            
            
            mp[flight_id].append(random.choice(class1))
            mp[flight_id].append(str(random.choice([0,1])))

            
mp
```

```python
lst=[]
for i in range(300):
    user=random.randint(1,25)
    route1=random.choice(mp2)
    flight_id=random.choice(mp4[route1])
    airline1=flight_id.split(":")[2]
    departure_time_temp=mp[flight_id][0]

    if 5 <= departure_time_temp.hour < 12:
        departure_time="05-12"
    elif 12 <= departure_time_temp.hour< 18:
        departure_time= "12-18"
    elif 18 <=departure_time_temp.hour< 23:
        departure_time= "18-23"
    else:
        departure_time= "23-05"

    class2=mp[flight_id][2]
    meal=mp[flight_id][3]

    lst.append([user,route1,airline1,flight_id,class2,meal,departure_time]) 

lst
```

```python
df = pd.DataFrame(lst, columns=['user','location','airline','flight_id','class','meal',"departure_time"])
```

```python
df.head()
```

```python
result = df.groupby('user').agg({
    'location': ','.join,
    'airline': ','.join,
    'flight_id': ','.join,
    'class': ','.join,
    'meal': ','.join,
    'departure_time': ','.join
}).reset_index()
```

```python
columns_to_encode = [col for col in df.columns if col not in ['user',"meal"]]
columns_to_encode
```

```python


for col in columns_to_encode:
    result[col] = result[col].str.split(',')
    
    unique_categories = set(category for sublist in result[col] for category in sublist)

    # Initialize new columns with zeros for each unique category
    for category in unique_categories:
        result[category] = 0

    for idx, row in result.iterrows():
        for category in row[col]:
            result.at[idx, category] += 1

result.head()
```

```python
result['meals'] = result['meal'].apply(lambda x: sum(int(value) for value in x.split(',')))
# sum1=sum(result['stop'])
# result['stop']=result['stop'].apply(lambda x:x/(sum1))
result=result.drop(['location','airline','flight_id','meal','class','departure_time'],axis=1)
result['meals']=result['meals']/result['meals'].max()
result.head()
```
